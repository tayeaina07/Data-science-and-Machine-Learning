```python
## getting an error so i saw I have to rename the gdp_per year ($) becuase there is a space before the dollar sign

df=df.rename(columns={'country':'Country','year':'Year','sex':'Gender','age':'Age'
                     ,'suicides_no':'SuicidesNo','population':'Population'
                      ,'gdp_for_year ($) ':'GdpForYear($)','population/100k' : 'Population/100k', 'suicides_no/100k' : 'Suicides_no/100k'})

```

```python
df.columns = df.columns.str.replace(" ", "_")
df
```

```python
for ind, row in df.iterrows():
    df.loc[ind, 'gdp_per_capital'] = row['_gdp_for_year_($)_']/ row['Population']
```

```python
plt.figure(figsize=(12,5))
df = df.groupby(['Year','Gender'])[['SuicidesNo']].sum().reset_index()
df['SuicidesNo'] = df['SuicidesNo']
ax = sns.lineplot(x="Year", y="SuicidesNo",hue = 'Gender',data=df,  legend="full")
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Suicide Rate by Gender ')
plt.show()
```

## question 12

```python
## to make by age visualization we much make the age into categorize into dummies.

df['Age'] = df['Age'].replace('5-14 years', 0)
df['Age'] = df['Age'].replace('15-24 years', 1)
df['Age'] = df['Age'].replace('25-34 years', 2)
df['Age'] = df['Age'].replace('35-54 years', 3)
df['Age'] = df['Age'].replace('55-74 years', 4)
df['Age'] = df['Age'].replace('75+ years', 5)
```

```python
# suicides in different age groups

x1 = df[df['Age'] == 0]['SuicidesNo'].sum()
x2 = df[df['Age'] == 1]['SuicidesNo'].sum()
x3 = df[df['Age'] == 2]['SuicidesNo'].sum()
x4 = df[df['Age'] == 3]['SuicidesNo'].sum()
x5 = df[df['Age'] == 4]['SuicidesNo'].sum()
x6 = df[df['Age'] == 5]['SuicidesNo'].sum()

x = pd.DataFrame([x1, x2, x3, x4, x5, x6])
x.index = ['5-14', '15-24', '25-34', '35-54', '55-74', '75+']
x.plot(kind = 'bar', color = 'blue')

plt.title('suicides in different age groups')
plt.xlabel('Age Group')
plt.ylabel('count')
plt.show()
```

```python
## it is clear from the above graphy that the age group 35-54 commit most suicide follwed by 55-74.
```
